Full Spectra:

```python
def ps(x,fs=1):
    X = np.fft.fft(x)/fs
    N=len(x)
    T = N/fs
    Gxx = X*np.conj(X)*2/T
    Gxx = np.abs(Gxx[:N/2+1])
    f = np.linspace(0,N/2+0.2,N/2+1.)/T
    return Gxx,f 
```

```python
GxxDC,fDC = ps(dc[1])
GxxJB,fJB = ps(jb[1])
GxxJM,fJM = ps(jm[1])
GxxISC,fISC = ps(isc[1])

Gxx = np.array([[fDC,fJB,fJM,fISC],[GxxDC,GxxJB,GxxJM,GxxISC]])
```

```python
def RemoveRowsOfNans(data):
    count = 0
    for i in data:
        if np.isnan(i).all(0):
            count += 1
    AllNansLoc = np.zeros(count)
    count = 0
    for ind, i in enumerate(data):
        if np.isnan(i).all(0):
            AllNansLoc[count] = ind
            count += 1
    data = np.delete(data, AllNansLoc, axis=0)
    return data
```

```python
def ReplaceNans(data):
    ReplacedData = 1.*data
    for ind, a in enumerate(data):
        where_are_NaNs = np.isnan(a)
        a[where_are_NaNs] = np.mean(a[~np.isnan(a)])
        ReplacedData[ind] = a
    return ReplacedData
```
