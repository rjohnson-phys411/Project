Spectra:

Full Spectra:

```python
def ps(x,fs=1):
    X = np.fft.fft(x)/fs
    N=len(x)
    T = N/fs
    Gxx = X*np.conj(X)*2/T
    Gxx = np.abs(Gxx[:N/2+1])
    f = np.linspace(0,N/2+0.2,N/2+1.)/T
    return Gxx,f 
```

```python
GxxDC,fDC = ps(dc[1])
GxxJB,fJB = ps(jb[1])
GxxJM,fJM = ps(jm[1])
GxxISC,fISC = ps(isc[1])
```

```python
nfft=1024*50
args={'NFFT':nfft,'Fs':1.,'window':mlab.window_hanning,'noverlap':nfft/2}

ppDC,ffDC = mlab.psd(dc[1],**args)
ppJB,ffJB = mlab.psd(jb[1],**args)
ppJM,ffJM = mlab.psd(jm[1],**args)
ppISC,ffISC = mlab.psd(isc[1],**args)

pDCJB,f=mlab.csd(dc[1],jb[1],**args)
pDCJM,f=mlab.csd(dc[1],jm[1],**args)
pDCISC,f=mlab.csd(dc[1],isc[1],**args)
pJBJM,f=mlab.csd(jb[1],jm[1],**args)
pJBISC,f=mlab.csd(jb[1],isc[1],**args)
pJMISC,f=mlab.csd(jm[1],isc[1],**args)


gammasq_DCJB = abs(pDCJB)**2/ppDC/ppJB
gammasq_DCJM = abs(pDCJM)**2/ppDC/ppJM
gammasq_DCISC = abs(pDCISC)**2/ppDC/ppISC
gammasq_JBJM = abs(pJBJM)**2/ppJB/ppJM
gammasq_JBISC = abs(pJBISC)**2/ppJB/ppISC
gammasq_JMISC = abs(pJMISC)**2/ppJM/ppISC
```

```python
def plotCompareInterp(xb,yb,Zb,titl='Binned Result'):
    Fontsize=12
    Labelsize=12
    Length=10
    Linewidth=0.5
    Alpha=0.8
    
    fig,ax=plt.subplots(1,1,figsize=(6,6))
    
    pcm=ax.pcolormesh(xb,yb,ma.masked_invalid(Zb),cmap=cm.RdBu_r)
    pcm.set_clim([np.min(Zb),np.max(Zb)])
    plt.colorbar(pcm,ax=ax,shrink=0.5)
    #cnt=ax.contour(x,y,z,colors='k')

    ax.set_aspect(1);ax.set_xlabel('x [km]');ax.set_ylabel('y [km]');ax.set_title(titl)
    
    ax.plot(StatLat,StatLong,'o',color='k') #location of stations with data
    ax.plot(noData[0],noData[1],'o',color='r') #location of stations without data

    ax.plot(MapLat,MapLong,'k',label='Vancouver Island',linewidth=Linewidth+0.5,alpha=Alpha)
    
    
    ax.set_xlim(np.min(xb),np.max(xb))
    ax.set_ylim(np.min(yb),np.max(yb))

    fmt=mticker.ScalarFormatter(useOffset=False)
    fmt.set_scientific(False)
    ax.xaxis.set_major_formatter(fmt)
    plt.tight_layout()
```
